```python
import pandas as pd

# Membaca file CSV dan mengonversi kolom 'Date' ke format tanggal
df = pd.read_csv('sales_data.csv', parse_dates=['Date'])

# Menampilkan beberapa baris pertama data
print(df.head())



```

```python
# Menambahkan kolom Total Revenue dengan mengalikan Order_Quantity dengan Unit_Price
df['Total Revenue'] = df['Order_Quantity'] * df['Unit_Price']

# Mengelompokkan data berdasarkan 'Product' dan menjumlahkan Total Revenue
total_revenue_per_item = df.groupby('Product')['Total Revenue'].sum().reset_index()

# Menampilkan hasil
print(total_revenue_per_item)
```

```python
# Menghitung profit margin untuk setiap order
df['Profit_Margin'] = df['Profit'] / df['Revenue']

# Memfilter order dengan profit margin lebih besar dari 0,2
high_margin_orders = df[df['Profit_Margin'] > 0.2]

# Menampilkan hasil
print(high_margin_orders)
```

```python
# Menghitung profit margin untuk setiap order
df['Profit_Margin'] = df['Profit'] / df['Revenue']

# Identifikasi produk dengan profit margin lebih kecil dari 0,1
low_margin_products = df['Profit_Margin'] < 0.1

# Terapkan diskon 10% pada Unit_Price untuk produk dengan profit margin kecil
df.loc[low_margin_products, 'Unit_Price'] = df['Unit_Price'] * 0.9

# Menampilkan hasil
print(df.head())  # Menampilkan beberapa baris pertama dari data yang telah dimodifikasi
```

```python
# Menambahkan kolom Total Revenue dengan mengalikan Order_Quantity dengan Unit_Price
df['Total Revenue'] = df['Order_Quantity'] * df['Unit_Price']

# Mengelompokkan data berdasarkan 'Product_Category' dan menjumlahkan Total Revenue
total_revenue_per_category = df.groupby('Product_Category')['Total Revenue'].sum().reset_index()

# Menampilkan hasil
print(total_revenue_per_category)
```

```python
# Menambahkan kolom Total Revenue dengan mengalikan Order_Quantity dengan Unit_Price
df['Total Revenue'] = df['Order_Quantity'] * df['Unit_Price']

# Mengelompokkan data berdasarkan 'Age_Group' dan menghitung rata-rata Total Revenue
average_revenue_per_age_group = df.groupby('Age_Group')['Total Revenue'].mean().reset_index()

# Menampilkan hasil
print(average_revenue_per_age_group)
```

```python
# Menyaring data untuk tahun tertentu (misalnya 2014)
df_2014 = df[df['Date'].dt.year == 2014]

# Menyimpan data yang telah difilter ke file CSV baru
df_2014.to_csv('sales_data_2014.csv', index=False)

# Menampilkan beberapa baris pertama dari data yang telah difilter
print(df_2014.head())
```

```python
# Menambahkan kolom Total Revenue dengan mengalikan Order_Quantity dengan Unit_Price
df['Total Revenue'] = df['Order_Quantity'] * df['Unit_Price']

# Mengelompokkan data berdasarkan 'Country' dan menjumlahkan Total Revenue
revenue_by_country = df.groupby('Country')['Total Revenue'].sum()

# Import the matplotlib library and alias it as plt
import matplotlib.pyplot as plt

# Membuat pie chart
plt.figure(figsize=(10, 7))
plt.pie(revenue_by_country, labels=revenue_by_country.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales by Country')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Menampilkan pie chart
plt.show()
```

```python
# Plot histogram dari distribusi usia customer
plt.figure(figsize=(10, 6))
plt.hist(df['Customer_Age'], bins=15, edgecolor='black', color='skyblue')

# Menambahkan label dan judul
plt.xlabel('Customer Age')
plt.ylabel('Frequency')
plt.title('Histogram of Customer Age Distribution')

# Menampilkan histogram
plt.grid(True)
plt.show()
```

```python
# Menambahkan kolom 'Month' dan 'Year' dari kolom 'Date'
df['Month'] = df['Date'].dt.to_period('M')
df['Year'] = df['Date'].dt.year

# Menghitung total revenue per bulan
monthly_revenue = df.groupby('Month')['Total Revenue'].sum()

# Membuat grafik garis dari revenue bulanan
plt.figure(figsize=(12, 6))
monthly_revenue.plot(kind='line', marker='o', color='teal')

# Menambahkan label dan judul
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Monthly Revenue')

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```
